```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
import math
```

Pre-processing data

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('crop_yield.csv')
df = df[df["Crop"] == "Rice"]
df
```

```python
#converting to per unit area

df["Fertilizer/Area"] = df["Fertilizer"]/df["Area"]
df["Pesticide/Area"] = df["Pesticide"]/df["Area"]

df
```

```python
#extracting only useful features

df = df[["Crop_Year", "Season", "State", "Annual_Rainfall", "Fertilizer/Area", "Pesticide/Area", "Yield"]]
df
```

```python
##converting rabi/kharif to summer/winter/autumn
```

```python
##one hot encoding Season and State features


columns_to_encode = ["Season", "State"]

df = pd.get_dummies(df, columns = columns_to_encode)


```

```python
#splitting dataset for training and testing

df_train = df[1:900]
df_test = df[901:]

X_train = df_train.drop(columns="Yield")
y_train = df_train["Yield"]

X_test = df_test.drop(columns="Yield")
y_test = df_test["Yield"]
```

Random Forests Library


```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, y_train)

```

```python
from sklearn.metrics import accuracy_score, mean_squared_error

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
feature_importances = model.feature_importances_
print("Feature Importances:", feature_importances)
```
